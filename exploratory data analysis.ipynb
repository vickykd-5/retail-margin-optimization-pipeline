```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("SuperStore Sales DataSet.xlsx")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df['Returns'].fillna(0)
```

# ship mode

```python
df['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Ship Mode', data=df, hue='Category', palette='summer')
plt.title('count of Ship Mode')
plt.savefig("count of Ship Mode.jpg")
plt.show()
```

# segment

```python
plt.figure(figsize=(4,4))
sns.countplot(x='Segment', data=df)
plt.title('count of Customer Segment')
plt.show()
```

# Product Category

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Category', data=df[df["Category"]=="Furniture"], hue='Sub-Category')
plt.title('count of Ship Mode')
plt.show()
```

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
df['Order Year'].value_counts()
```

```python
sns.barplot(x="Category", y="Profit", data=df,estimator='sum')
plt.show()
```
